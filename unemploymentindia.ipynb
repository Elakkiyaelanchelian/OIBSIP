```python
X = 2.5 * np.random.randn(100) + 3
residuals = 0.5 * np.random.randn(100)
y = 2 + 0.3 * X + residuals
df = pd.DataFrame({'X': X, 'y': y})
print(df.isnull().sum())
X_train, X_test, y_train, y_test = train_test_split(df[['X']], df['y'])
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_test['X'], y=y_test, color='blue', label='Actual')
sns.scatterplot(x=X_test['X'], y=y_pred, color='red', label='Predicted')
plt.plot(X_test, y_pred, color='green', linewidth=2, label='Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Results')
plt.legend()
plt.show()
```
